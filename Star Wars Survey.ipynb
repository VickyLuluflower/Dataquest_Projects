# Star Wars Survey
#### Project Goal:
To answer the question:`does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?`
#### Project Overview:
To address this question, we surveyed Star Wars fans and received 835 total responses, which is available from [their GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey)

```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

```python
# Remove invalid data from the 1st row of the 9th column
star_wars = star_wars.drop([0])
star_wars.head(3)
```

```python
star_wars.columns
```

## Cleaning and Mapping Yes/No Columns

Take a look at the next two columns:

- `Have you seen any of the 6 films in the Star Wars franchise?`
- `Do you consider yourself to be a fan of the Star Wars film franchise?`
Both represent `Yes/No` questions. They can also be `NaN` where a respondent chooses not to answer a question. 

We'll convert the two columns to the Boolean type.

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

## Cleaning and Mapping Checkbox Columns


```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"    
})
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

## Cleaning the Ranking Columns

```python
# Rename the columns to something readable
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python
# Convert all values to floats. (Leave NaN as is.)
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

## Finding the Highest-Rank Movie

```python
# Compute the mean of each ranking columns 
rankings_mean = star_wars[star_wars.columns[9:15]].mean()
print(rankings_mean)
```

```python
# Make a bar chart for each seen
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), rankings_mean)
plt.title('Average Ranking for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Ranking')
plt.show()
```

Episode V is the most favorite movie, as lower numbers means higher rankings.

The first trilogy (episodes IV, V, VI) is more popular than the second trilogy (episodes I, II, III).

## Finding the Most Viewed Movie

```python
# Compute the sum of each seen columns
seens_sum = star_wars[star_wars.columns[3:9]].sum()
print(seens_sum)
```

```python
# Make a bar chart of each ranking columns
plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), seens_sum)
plt.title('Average Number of Views for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Number of Views')
plt.show()
```

What we see that episode V is not only the most popular one in terms of ranking, but also the one that has been seen most. Followed by episode VI, also both for ranking and number of people who watched.

It seems that the first trilogy (episodes IV, V, VI) are most popular.

## Exploring the Data by Binary Segments

Now let's examine how certain segments of the survey population responded. There are several columns that segment our data into two groups. Here are a few examples:

- Do you consider yourself to be a fan of the Star Wars film franchise? — `True` or `False`
- Do you consider yourself to be a fan of the Star Trek franchise? — `Yes` or `No`
- Gender — `Male` or `Female`

We can split a DataFrame into two groups based on a binary column by creating two subsets of that column. 

### Gender Differences

```python
# Split on the Gender column into two groups
females = star_wars[star_wars['Gender']=='Female']
males = star_wars[star_wars['Gender']=='Male']
```

```python
# Compute the most viewed movie for females
female_seens_sum = females[females.columns[3:9]].sum()
print(female_seens_sum)
```

```python
# Make a bar chart of each ranking columns
plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), female_seens_sum)
plt.title('Female: Total Number of Views for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Number of Views')
plt.show()
```

```python
# Compute the highest-ranked movie for female
female_rankings_mean = females[females.columns[9:15]].mean()
print(female_rankings_mean)
```

```python
# Make a bar chart of each ranking columns
plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), female_rankings_mean)
plt.title('Female: Average Rankings for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Number of Views')
plt.show()
```

```python
# Compute the most viewed movie for males
male_seens_sum = males[males.columns[3:9]].sum()
print(male_seens_sum)
```

```python
# Make a bar chart of each ranking columns
plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), male_seens_sum)
plt.title('Male: Total Number of Views for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Number of Views')
plt.show()
```

```python
# Compute the highest-ranked movie for males
male_rankings_mean = males[males.columns[9:15]].mean()
print(male_rankings_mean)
```

```python
# Make a bar chart of each ranking columns
plt.bar(('Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI'), male_rankings_mean)
plt.title('Male: Average Rankings for each Episode', fontsize=16)
plt.xlabel('Star Wars Movie by Episode')
plt.ylabel('Average Number of Views')
plt.show()
```

```python
# Make a grouped bar chart to compare gender differences in Number of Views

labels = ['Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

plt.bar(x - width/2, female_seens_sum, width, label='Females')
plt.bar(x + width/2, male_seens_sum, width, label='Males')

# Add some text for labels and title 
plt.title('Total Number of Views by Gender')
plt.ylabel('Total Number of Views')
plt.xlabel('Star Wars Movie by Episode')
plt.legend()

plt.show()
```

Both males and females like Episodes IV, V and VI most. However, men seem to be more outspoken about that than females

```python
# Make a grouped bar chart to compare gender differences in Average Rankings

labels = ['Ep I', 'Ep II', 'Ep III', 'Ep IV', 'Ep V', 'Ep VI']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

plt.bar(x - width/2, female_rankings_mean, width, label='Females')
plt.bar(x + width/2, male_rankings_mean, width, label='Males')

# Add some text for labels and title 
plt.title('Average Rankings by Gender')
plt.ylabel('Average Rankings')
plt.xlabel('Star Wars Movie by Episode')
plt.legend()

plt.show()
```

All episodes were watched more by men then women.
More males watches episodes 1-3, but males liked them far less than females did.

- The first trilogy (episodes IV, V, VI) is generally more popular that the second trilogy (episodes I, II, III)
