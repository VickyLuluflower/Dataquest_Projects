# Storytelling Data Visualization on Exchange Rates
In this project, we'll explore the [Daily Exchange Rates per Euro 1999-2021](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) dataset. 

## Introducing the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

## Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar', 
                               'Period\\Unit:':'Time'}, 
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() 
```

62 has '-' character

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Mean
Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
# Enables Jupyter to display graphs:
%matplotlib inline 

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

If we want to hide the daily variation on the graph and show only the long-term trends, we can use the rolling mean (also known as the moving average). 
<br>
We calculated the rolling mean using values from two days — the current day and the previous day. We call the number of days used in the calculation the rolling window (or moving window). 
<br>
The window gets larger, and the line becomes smoother. This increases the data-ink ratio, and it can be useful if we want the audience to focus only on long-term trends.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar['rolling_mean']
```

```python
plt.plot(euro_to_dollar['rolling_mean'])
plt.show()
```

## Coming up with an Idea
Now we're going to spend some time coming up with an idea for the graph we want to build. Our goal is to create a storytelling data visualization using the data we have.

To create a story, remember that we need to arrange our data into a series of events that show change.

We'll show how the euro-dollar rate has changed during the coronavirus pandemic. 
We can show the 2020 data and the 1999-2019 data as a baseline. We can use a line plot.

```python
covid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year == 2020)]
before_covid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 1999)
                                     &(euro_to_dollar['Time'].dt.year <= 2019)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Highlighting the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(covid['Time'], covid['rolling_mean'],
       linewidth=3, color='red')

### Adding the 1999-2019 period
ax.plot(before_covid['Time'], before_covid['rolling_mean'],
       linewidth=1, color='green')

### Adding a title and a subtitle
ax.set_title('Euro-USD rate increased to 1.20 during Coronavirou Pandemic',
       size=15, weight='bold')

### Adding some transparency to grid
ax.grid(alpha=0.2)

plt.show()
```
